```python
punct = ['. ', '! ','? ', '؟ ', ': ', '?! ']
```

```python
def split(poems, punct):
    ids = []
    splitted_poems = [(poem.split(punct)) for poem in poems]
    list_sentences = []
    for splitted_poem in splitted_poems:
        if len(splitted_poem) > 1:
            for i in range(len(splitted_poem) - 1):
                new_sent = splitted_poem[i] + punct
                list_sentences.append(new_sent)

            list_sentences.append(splitted_poem[-1])
        
        else:
            list_sentences.append(splitted_poem[0])
        
        
            
    return list_sentences
```

```python
new_list = poems_strings.copy()
for pun in punct:
    new_list = split(new_list, pun)

new_list = [(sent) for sent in new_list if len(sent) > 1]
```

```python
import ufal.udpipe
```

```python
model = ufal.udpipe.Model.load('/Users/kseniapetuhova/Desktop/arabic-padt-ud-2.5-191206.udpipe')
tokenizer = model.newTokenizer(model.DEFAULT)
conlluOutput = ufal.udpipe.OutputFormat.newOutputFormat("conllu")
sentence = ufal.udpipe.Sentence()
error = ufal.udpipe.ProcessingError()
```

```python
def tag(text, model):
    tokenizer.setText(text)
    tokenizer.nextSentence(sentence, error)
    model.tag(sentence, model.DEFAULT)
    model.parse(sentence, model.DEFAULT)
    return conlluOutput.writeSentence(sentence)
```
